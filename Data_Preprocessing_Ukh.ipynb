# Зависимость между уровнем дохода и возвратом кредита в срок

```python
data['monthly_income'].quantile([0.25,0.5,0.75])
```

Распределим уровень дохода на 5 категорий

```python
def total_income_category(monthly_income):
    if monthly_income <= 30000:
            return 'E'
    if monthly_income <= 50000:
            return 'D'
    if monthly_income <= 200000:
            return 'C'
    if monthly_income <= 1000000:
            return 'B'
    #if monthly_income > 1000000:
           # return 'A'
    return 'A'
```

```python
data['total_income_category'] = data['monthly_income'].apply(total_income_category)
```

```python
data.head()
```

```python
monthly_income_set = data.groupby('total_income_category')['credit_repayment_status'].sum() / \
      data.groupby('total_income_category')['credit_repayment_status'].count() * 100
```

```python
monthly_income_set.sort_values()
```

```python
monthly_income_pivot = data.pivot_table(index=['total_income_category'], columns=['credit_repayment_status'], values='education_level_id', aggfunc='count')
monthly_income_pivot['ratio'] = monthly_income_pivot[1] / monthly_income_pivot[0] * 100
monthly_income_pivot
```

# Как разные цели кредита влияют на его возврат в срок?

Используя map(), составим словарь из двух списков, а потом сошьем.

```python
credit_purpose_keys= []
credit_purpose_list = list(data.credit_purpose.unique())
for value in range(len(credit_purpose_list)):
    if 'образован' in credit_purpose_list[value]:
        credit_purpose_keys.append('Получение образования')
    elif 'авто' in credit_purpose_list[value]:
        credit_purpose_keys.append('Операции с автомобилем')
    elif 'свадьб' in credit_purpose_list[value]:
        credit_purpose_keys.append('Проведение свадьбы')
    else:
        credit_purpose_keys.append('Операции с недвижимостью')
        
credit_purpose_keys
```

```python
credit_purpose_pivot = data.pivot_table(index=['reason'], columns=['credit_repayment_status'], values='education_level_id', aggfunc='count')
credit_purpose_pivot['ratio'] = credit_purpose_pivot[1] / credit_purpose_pivot[0] * 100
credit_purpose_pivot
```
